Linear Combinations of Sine and cosine.<br>
$$a\cos(\omega \theta)+b\sin(\omega \theta)=A(\sin(\omega \theta+\phi))\tag{eq 1}$$<br>
On the right side we use the identity $$\sin(\alpha+\beta)=\sin\alpha\cos\beta + \cos\alpha\sin\beta$$ to get
$$a\color{red}{\cos(\omega \theta)}+b\color{blue}{\sin(\omega \theta})=A(\color{blue}{\sin\omega\theta}\cos\phi+ \color{red}{\cos\omega\theta}\sin\phi)$$ from which, we can rearrange the terms
    $$a\cos(\omega \theta)+b\sin(\omega\theta)
    =(A\sin\phi)\color{red}{\cos\omega\theta} + (A\cos\phi)\color{blue}{\sin\omega\theta}$$<br>
    Now we can set equal the coefficients of $\cos\omega\theta$ and $\sin\omega\theta$<br>
    $$a=A\sin\phi \qquad b=A\cos\phi$$<br>
    If we imagine a right triangle, then it is clear that 'a' and 'b' are the leg lengths and 'A' is the hypotenuse.
    <img src="../Images/RightTriangle1.png">
    Since we know 'a' and 'b', we can use Pythagoras to compute 'A' and we can use $tan\phi=\frac{a}{b}$ to find $\phi.$

<font size=3>
        In the plot below we let $a=3$ , $b=4$ and $\omega=1$. Using the Pythagorean triangle, we find that $A=5$ and that $\phi=tan^{-1}(3/4).$ We plot both sides on the graph and see that they overlay.
    $$y=3\cos(t)+4\sin(t)$$ $$y_1=5\sin\left(t+tan^{-1}(3/4)\right)$$

```python
%reset -f
from numpy import *
import matplotlib.pyplot as plt
%matplotlib inline

x=linspace(0,2*pi,num=101)
y=3*cos(x)+4*sin(x); #A=5, Ï•=arctan(3/4) 

y1 = 5*sin(x+arctan(3/4))

fig,ax=plt.subplots(figsize=(5,5))

plt.plot(x,y,linewidth=3)
plt.plot(x,y1)

plt.title('Linear Combination of Cos and Sin',size=20)
```

<font size=3>
    We can also create the combination with cosine on the right, but notice that we have changed the sign inside the cosine.
    $$a\cos(t)+b\sin(t)=A(cos(t-\phi))$$<br>
    Again use the identity to expand the right side: $$cos(\alpha-\beta)=cos\alpha\cos\beta+sin\alpha\sin\beta$$<br>
    $$a\color{red}{\cos t}+b\color{blue}{\sin t} =A\color{red}{\cos t}\cos\phi + A\color{blue}{\sin t} \sin\phi$$<br>
    $$a\color{red}{\cos t}+b\color{blue}{\sin t} =A\cos\phi\color{red}{\cos t}+A\sin\phi\color{blue}{\sin t}$$<br>
    So matching coefficients we get
    $$a=A\cos\phi\qquad b=A\sin\phi$$<br>
    We can use the same Pythagoras relation as before.  What has changed is the sign inside the cosine on the right and also the definitions of $a$ and $b.$
   <img src="../Images/RightTriangle.png">
    This time, $tan\phi=\frac{b}{a}$
    

```python
%reset -f
from numpy import *
import matplotlib.pyplot as plt
%matplotlib inline

x=linspace(0,2*pi,num=101)
y=3*cos(x)+4*sin(x); #A=5, Ï•=arctan(3/4) 

y1 = 5*cos(x-arctan(4/3))

fig,ax=plt.subplots(figsize=(5,5))

plt.plot(x,y,linewidth=3)
plt.plot(x,y1)

plt.title('Linear Combination of Cos and Sin',size=20)
```
