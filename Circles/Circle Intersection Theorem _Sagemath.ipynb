Let's look at the intersection of a circle and a line or you may think of it as two circles. The point is that we can create a unique circle that passes thru the intersections of the two objects. This will illustrate the general idea in a simple context.

The proof is elementary but needs a few prerequisites:

(1) The general equation of a circle is $ax^2+ay^2+bx+cy+d=0, a\ne 0$.<br>A second degree equation in $x$ and $y$ with equal
(non-zero) coefficients of $x^2$ and $y^2$ and with no terms in $xy$ is a circle.

(2) The general equation of a line is $lx+my+n=0$.

(3) There is precisely one circle through any given three non-collinear points.

**Theorem** 

Let $f(x,y)=0$ be the equation of a circle $\mathcal{C}$ and let $g(x,y)=0$ be the equation of a line $\mathcal{L}$ or another circle, which intersects $\mathcal{C}$ at two points, $P$ and $Q$. 

(a) For a constant $k$, $f(x,y)+kg(x,y)=0$ is the equation of a circle through $P$ and $Q$. 

(b) Every circle through $P$ and $Q$ has  equation $\color{red}{f(x,y)+kg(x,y)=0}$  for some constant $k$. 

**Proof**

(a) From prerequisites (1) and (2), $f(x,y)+kg(x,y)=0$ is the equation of a circle, $\mathcal{D}$ say. 

Now let $(a,b) $ be the coordinates of either $P$ or $Q$, then $f(a,b)=g(a,b)=0$.  Therefore $f(a,b)+kg(a,b)=0$ and so $(a,b)$ is a point on $\mathcal{D}$. Thus (a) is proved.

(b) Let $\mathcal{E}$ be any circle through $P$ and $Q$. Let $R(c,d)$ be a third point on $\mathcal{E}$ and then $R$ is automatically not on the line $\mathcal{L}$.

Since $R$ is not on $\mathcal{L}$, $g(c,d)\ne 0$. We can therefore find $k$ such that  $f(c,d)+kg(c,d)=0$. Then $f(x,y)+kg(x,y)=0$ represents a circle through $P,Q$ and $R$ and, by prerequisite (3), it is the circle  $\mathcal{E}$, as required.

```python
var('x y k');D=[];
eq1=x^2+y^2-3*x+4*y-1==0;
eq2=(x+2)^2+(y+2)^2==4;
for k in range(-5,14,1):
    D.append( (x^2+y^2-3*x+4*y-1) + k*((x+2)^2+(y+2)^2-4)==0 );

## Create a plot object
a = plot ([] , figsize =(5)) #be careful about reusing variables like this!!!!

## Set the plot parameters
xmin = -5; xmax = 5 ; ymin = -5; ymax = 1 ## Graph Bounds
for k in range(-10,14,1):
    a += implicit_plot(D[k],(xmin,xmax),(ymin,ymax),color='red',linewidth=.5)
a += implicit_plot (eq1,(xmin,xmax),(ymin,ymax), color ='green', title ="Circles on the 2 intersections", fontsize =12)
a += implicit_plot(eq2,(xmin,xmax),(ymin,ymax),color='blue')

#a += list_plot (B, color ='green', size =1)
#a += plot (f ,(0 , tEnd ), thickness =5, alpha =.3) # opacity ( alpha ) is reduced to show overlap
a. set_axes_range (xmin ,xmax ,ymin , ymax )
#a. axes_labels (['$ time ( seconds )$',r'$Conc. \mu Molar$'])
#a. fontsize (10)
a. axes ( False )
show (a, frame = True )
```

The sum of any two circles is a circle.  If the sum results in a negative radius, it will be an imaginary circle. Let's represent the circle(s) as $(x-h)^2+(y-k)^2=r^2$ Then we can show equations for two circles expanded.
$$x^2+y^2+h_1x-2k_1y+h_1^2+k_1^2-r_1^2=0$$
$$x^2+y^2+h_2x-2k_2y+h_2^2+k_2^2-r_2^2=0$$
Now we can add them together: with $consts=h_1^2+h_2^2+k_1^2+k_2^2-r_1^2-r_2^2$
$$2x^2+2y^2-2(h_1+h_2)-2(k_1+k_2)y+consts=0$$
