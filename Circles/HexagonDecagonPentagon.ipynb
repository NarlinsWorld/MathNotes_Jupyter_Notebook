The same radius circle can be circumscribed around each of the polygons below.

```python
from numpy import exp, pi, arange, array
import matplotlib.pyplot as plt
%matplotlib inline
    
# Vertices of a hexagon
hexa = exp(2j*pi*arange(6)/6)

# Vertices of a decagon, shifted and rotated
# to have a side perpendicular to the hexagon
deca = exp(2j*pi*(0.5+arange(10))/10)
deca += hexa[1] - deca[5]
print(hexa[1])

# Shift and rotate a pentagon to share one vertex 
# with the hexagon and one vertex with the decagon
omega = exp(2j*pi/5)
c = (hexa[2]*omega - deca[4])/(omega - 1)
penta = c + (hexa[2] - c)*exp(2j*pi*arange(5)/5) 

def connect(p, q, c):
    plt.plot([p.real, q.real], [p.imag, q.imag], '-', color=c)
plt.gca().set_aspect("equal")

D=[] #deca
P=[] #penta
for n in range(-1, 5):
    connect(hexa[n], hexa[n+1], 'blue')
for n in range(-1, 9):
    connect(deca[n], deca[n+1], 'green')
    D.append([deca[n].real,deca[n].imag])
for n in range(-1, 4):
    connect(penta[n], penta[n+1], 'red')
    P.append([penta[n].real,penta[n].imag])
    #print(P[n])

#Center & radius Coordinates of Circle 1:
r_1=1.0
A = array((0, 0))
circle1 = plt.Circle(A, r_1, color='b', fill=False)
plt.gca().add_patch(circle1)

#print('D[1]', D[1])
#print('D[6]', D[6])
DC=[ (D[1][0]+D[6][0])/2, (D[1][1]+D[6][1])/2 ]
#print('DC',DC)
A = array((DC[0], DC[1]))
circle2 = plt.Circle(A, r_1, color='b', fill=False)
plt.gca().add_patch(circle2)

x_1=P[0][0]; x_2=P[1][0]; x_3=P[2][0]
y_1=P[0][1]; y_2=P[1][1]; y_3=P[2][1]
#degug
# plt.plot(x_1,y_1,marker="o", markersize=10, markeredgecolor="green", markerfacecolor="red", linestyle='none')
# plt.plot(x_2,y_2,marker="o", markersize=10, markeredgecolor="green", markerfacecolor="red", linestyle='none')
# plt.plot(x_3,y_3,marker="o", markersize=10, markeredgecolor="green", markerfacecolor="red", linestyle='none')
ED_x=(x_2-x_1); ED_y= y_2-y_1   #slope of [x_1,y_1] to [x_2,y_2] line
DC_x = (x_3-x_2); DC_y=y_3-y_2 #slope of [x_2,y_2] to [x_3,y_3] line
u_x = -ED_y; u_y=ED_x       #slope of the perp line from M1
v_x = -DC_y; v_y = DC_x      #slope of the perp line from M2
up_x = -u_y; up_y = u_x      #u perp

M1_x=(x_2+x_1)/2; M1_y = (y_2+y_1)/2
M2_x=(x_3+x_2)/2; M2_y = (y_3+y_2)/2
#plt.plot(M1_x,M1_y,marker='o')       #plots a point at M1
p_x = M1_x + 1*(u_x); p_y = M1_y+1*u_y
#plt.plot(p_x, p_y, marker = 'o')     #plots a point at p
#plt.plot([M1_x,p_x],[M1_y,p_y]) #draws a line from M1 to p

#plt.plot(M2_x,M2_y,marker='o')    #plots a point at M2
q_x = M2_x +1*(v_x); q_y = M2_y+1*v_y
#plt.plot(q_x,q_y,marker='o')    #plots a point at q
#plt.plot([M2_x,q_x],[M2_y,q_y])  #draws a line from M2 to q


# plt.plot(M1_x,M1_y,marker="o", markersize=10, markeredgecolor="green", markerfacecolor="red", linestyle='none')
# plt.plot(M2_x,M2_y,marker="o", markersize=10, markeredgecolor="green", markerfacecolor="red", linestyle='none')
num = (M1_x-M2_x)*up_x+(M1_y-M2_y)*up_y
den = (v_x * up_x)+(v_y * up_y)
#print('s=',num/den)


X_x = M2_x + (num/den)*v_x
X_y = M2_y + (num/den)*v_y

#print(X_x,X_y)
A=array((X_x,X_y))
circle3=plt.Circle(A,r_1,color='b', fill=False)
plt.gca().add_patch(circle3)
```
