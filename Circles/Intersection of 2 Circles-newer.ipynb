<h1>Simple 'by hand' intersection of 2 circles</h1>
<font size=4> This is not the method programmed below!!!
This can be done without any trigonometry at all. Let the equations of the circles be
$$(x-x_1)^2 + (y-y_1)^2 = r_1^2, \tag{1}$$
$$(x-x_2)^2 + (y-y_2)^2 = r_2^2. \tag{2}$$
By subtracting $(2)$ from $(1)$ and then expanding, we in fact obtain a *linear* equation for $x$ and $y$; after a little rearranging it becomes
$$-2x(x_1 - x_2) - 2y(y_1 - y_2) = (r_1^2 - r_2^2) - (x_1^2 - x_2^2) - (y_1^2 - y_2^2).$$
(If the circles intersect, this is the equation of the line that passes through the intersection points.) This equation can be solved for one of $x$ or $y$; let's suppose $y_1 - y_2 \ne 0$ so that we can solve for $y$:
$$y = -\frac{x_1 - x_2}{y_1 - y_2} x + \dotsc. \tag{3}$$
 Substituting this expression for $y$ into $(1)$ or $(2)$ gives a quadratic equation in only $x$. Then the $x$-coordinates of the intersection points are the solutions to this; the $y$-coordinates can be obtained by plugging the $x$-coordinates into $(3)$.

<font size=4> Example

$$(x+2)^2+(y-3)^2=\left(\frac{11}{2}\right)^2 \tag{EQ 1}$$
$$(x-4)^2+(y+5)^2=6^2 \tag{EQ 2}$$
$$x^2+4x+y^2-6y=\frac{69}{4} \tag{EQ 1a}$$
$$x^2-8x+y^2+10y=-5 \tag{EQ 2a}$$

$$12x-16y=\frac{89}{4} \tag{EQ 3}$$
Solving for $y.$
   $$y=\frac{48x-89}{64}$$
Substitute into EQ 2.
$$x^2-8x+\left[  \frac{48x-89}{64}\right]^2+10\left[ \frac{48x-89}{64} \right]=-5$$
 and now solve the quadratic for $x.$
 $$x=\pm\frac{12\sqrt{5719}+331}{400} \approx 3.096\ or\ -1.441$$
  Now substitute each of these $x$ values into EQ3 to get the corresponding $y$ values.
   $$y=\pm\frac{9\sqrt{5719}-308}{400}$$
   $$(x_1,y_1)=(3.096,\,0.932)$$
   $$(x_2,y_2)=(-1.441,\,-2.472)$$
    

<h2>Intersection of 2 circles</h2>

$t_1=h^3 - h^2 h_1 - h h_1^2 + h k^2 - 2h k k_1 + h k_1^2 - h r_1^2 + h r_2^2 + h_1^3 + h_1 k^2 - 2h_1 k k_1 + h_1 k_1^2 + h_1 r_1^2 - h_1 r_2^2$
$t_1=h^{3} - h^{2} \; h_1 - h \; h_1^{2} + h \; k^{2} - 2 \; h \; k \; k_1 + h \; k_1^{2} - h \; r_1^{2} + h \; r_2^{2} + h_1^{3} + h_1 \; k^{2} - 2 \; h_1 \; k \; k_1 + h_1 \; k_1^{2} + h_1 \; r_1^{2} - h_1 \; r_2^{2} $<br>
$t_2 = -h^{4} + 4 \; h^{3} \; h_1 - 6 \; h^{2} \; h_1^{2} - 2 \; h^{2} \; k^{2} + 4 \; h^{2} \; k \; k_1 - 2 \; h^{2} \; k_1^{2} + 2 \; h^{2} \; r_1^{2}$<br>
$t_3=2 \; h^{2} \; r_2^{2} + 4 \; h \; h_1^{3} + 4 \; h \; h_1 \; k^{2} - 8 \; h \; h_1 \; k \; k_1 + 4 \; h \; h_1 \; k_1^{2} - 4 \; h \; h_1 \; r_1^{2} - 4 \; h \; h_1 \; r_2^{2}$<br>
$t_4 = - h_1^{4} - 2 \; h_1^{2} \; k^{2} + 4 \; h_1^{2} \; k \; k_1 - 2 \; h_1^{2} \; k_1^{2} + 2 \; h_1^{2} \; r_1^{2} + 2 \; h_1^{2} \; r_2^{2}$<br>
$t_5 = - k^{4} + 4 \; k^{3} \; k_1 - 6 \; k^{2} \; k_1^{2} + 2 \; k^{2} \; r_1^{2} + 2 \; k^{2} \; r_2^{2} + 4 \; k \; k_1^{3} - 4 \; k \; k_1 \; r_1^{2} - 4 \; k \; k_1 \; r_2^{2}$<br>

$t_6=- k_1^{4} + 2 \; k_1^{2} \; r_1^{2} + 2 \; k_1^{2} \; r_2^{2} - r_1^{4} + 2 \; r_1^{2} \; r_2^{2} - r_2^{4}$<br>
$denom = 2 \; h^{2} - 4 \; h \; h_1 + 2 \; h_1^{2} + 2 \; k^{2} - 4 \; k \; k_1 + 2 \; k_1^{2}$<br>

$t_7 = h^{2} \; k + h^{2} \; k_1 - 2 \; h \; h_1 \; k - 2 \; h \; h_1 \; k_1 + h_1^{2} \; k + h_1^{2} \; k_1 + k^{3} - k^{2} \; k_1 - k \; k_1^{2} - k \; r_1^{2} + k \; r_2^{2} + k_1^{3} + k_1 \; r_1^{2} - k_1 \; r_2^{2} $<br>


<br><br>

$$ x=\frac{t_1+(-k+k_1)\sqrt{t_2+t_3+t_4+t_5+t_6}}{denom} \quad
 y=\frac{t_7+(h-h_1)\sqrt{t_2+t_3+t_4+t_5+t_6}}{denom} $$<br><br>
 $$ x=\frac{t_1+(k-k_1)\sqrt{t_2+t_3+t_4+t_5+t_6}}{denom} \quad
 y=\frac{t_7+(-h+h_1)\sqrt{t_2+t_3+t_4+t_5+t_6}}{denom} $$

```python
#Circle Intersection
%reset -f

from numpy import *
import matplotlib.pyplot as plt
%matplotlib inline

def CircleIntersection(A,B,r_1,r_2):
    h=A[0]; k=A[1];  h_1=B[0]; k_1=B[1]
    denom = 2*h**2  -4*h*h_1  +2*h_1**2  +2*k**2  -4*k*k_1  +2*k_1**2
    t_6 = -k_1**4  +2*k_1**2*r_1**2  +2*k_1**2*r_2**2  -r_1**4  +2*r_1**2*r_2**2  -r_2**4
    t_5  =-k**4  +4*k**3*k_1  -6*k**2*k_1**2  +2*k**2*r_1**2  + 2*k**2*r_2**2 + 4*k*k_1**3 - 4*k*k_1*r_1**2 - 4*k*k_1*r_2**2
    t_4 = -h_1**4  -2*h_1**2*k**2  +4*h_1**2*k*k_1  -2*h_1**2*k_1**2  +2*h_1**2*r_1**2  +2*h_1**2*r_2**2
    t_3 = 2*h**2*r_2**2   +4*h*h_1**3  +4*h*h_1*k**2   -8*h*h_1*k*k_1  +4*h*h_1*k_1**2  -4*h*h_1*r_1**2  -4*h*h_1*r_2**2
    t_2 = -h**4  +4*h**3*h_1  -6*h**2*h_1**2  -2*h**2*k**2  +4*h**2*k*k_1  -2*h**2*k_1**2  +2*h**2*r_1**2 
    t_1 = h**3 - h**2 * h_1 - (h * h_1**2) + (h * k**2) - (((2 * h) * k) * k_1) + (h * k_1**2) - (h * r_1**2) + (h * r_2**2) + h_1**3 + (h_1 * k**2) - (((2 * h_1) * k) * k_1) + (h_1 * k_1**2) + (h_1 * r_1**2) - (h_1 * r_2**2)
    t_7 = (h**2 * k) + (h**2 * k_1) - (((2 * h) * h_1) * k) - (((2 * h) * h_1) * k_1) + (h_1**2 * k) + (h_1**2 * k_1) + k**3 - (k**2 * k_1) - (k * k_1**2) - (k * r_1**2) + (k * r_2**2) + k_1**3 + (k_1 * r_1**2) - (k_1 * r_2**2) 
    x_1 = (t_1+(-k+k_1)*sqrt(t_2 + t_3 + t_4 + t_5 + t_6))/denom
    y_1 = (t_7+(h-h_1)*sqrt(t_2 + t_3 + t_4 + t_5 + t_6))/denom
    x_2 = (t_1+(k-k_1)*sqrt(t_2 + t_3 + t_4 + t_5 + t_6))/denom
    y_2 = (t_7+(-h+h_1)*sqrt(t_2 + t_3 + t_4 + t_5 + t_6))/denom
    C=array((x_1,y_1))
    D=array((x_2,y_2))
    return C,D

h=-2; k=3; A=array((h,k))
h_1=4; k_1=-5; B=array((h_1,k_1))
r_1=5.5
r_2=6

C,D = CircleIntersection(A,B,r_1,r_2)
print("C=",C)
print("D=",D)

#  Draw the circles with their intersection
fig, ax = plt.subplots(figsize=(8,8))

ax.set( aspect=1)
ax.grid(True)

#Center & radius Coordinates of Circle 1:
circle1 = plt.Circle(A, r_1, color='r', fill=False)
ax.add_patch(circle1)
ax.plot(h,k,marker=".",mfc="k",mec="k")

#Center & radius Coordinates of Circle 2:
circle2 = plt.Circle(B, r_2, color='b', fill=False)
ax.add_patch(circle2)
ax.plot(h_1,k_1,marker=".",markerfacecolor="black",mec="black")

plt.plot(C[0], C[1], marker="o", markersize=10, markeredgecolor="green", markerfacecolor="red")
plt.plot(D[0], D[1], marker="o", markersize=10, markeredgecolor="red", markerfacecolor="green")

#fetch some reasonable axes limits
xmax = ceil(1.1*max(A[0]+r_1,B[0]+r_2))
xmin = floor(1.1*min(A[0]-r_1,B[0]-r_2))
ymax = ceil(1.1*max(A[1]+r_1,B[1]+r_2))
ymin = floor(1.1*min(A[1]-r_1,B[1]-r_2))
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])
```
