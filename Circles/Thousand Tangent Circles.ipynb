```python
from numpy import *
r_0 = 1000;                #Radius of circle 0
C_0 = (0,r_0);             #Center of original circle

r_1 = 1/1000;              #Radius of 1st small tangent circle
xC1 = sqrt((r_0+r_1)**2 - (r_1-r_0)**2);  #

print("(" + str(xC1)+"," + str(r_1) +")");  #C1
#----------
for n in range(2,800):
    xC2 = (-10*sqrt(10)*sqrt(r_1)-1000)*xC1/(r_1-1000)

    r_2 = xC1**2/(4*r_1 + 4000 + 80*sqrt(10)*sqrt(r_1))
    print(n,xC2,r_2)
    
    xC1=xC2
    r_1=r_2
    
#print(xC2,r_2)
```
