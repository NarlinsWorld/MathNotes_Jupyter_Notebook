```python
%reset -f

from numpy import *
import matplotlib.pyplot as plt
%matplotlib inline



fig, ax = plt.subplots(figsize=(8,8))
ax.set( aspect=1)

A=array((3,3))                     #center of circle A
r_1 = 1                          #radius of circle A
B=array((3+2*r_1,3))               #center of circle B
r_2 = r_1                          #radius of circle B


circle1 = plt.Circle(A, r_1, color='b', fill=False)
ax.add_patch(circle1)
circle2 = plt.Circle(B,r_2,color='b',fill=False)
ax.add_patch(circle2)

dist = linalg.norm(A-B)            #distance from center A to center B
print("A=",A)
print("B=",B)
θ=pi/3                             #60 degrees
v=array((cos(θ),sin(θ)))           #vector pointing 60 degrees
C=A+dist*v                         #the new point in vector notation
r_3=linalg.norm(C-A)/2             #radius of circle C
print("C=",C)

circle3 = plt.Circle(C,r_3,color='b',fill=False)
ax.add_patch(circle3)

x=[A[0],B[0],C[0]]
y=[A[1],B[1],C[1]]
plt.plot(x,y,marker="o", markersize=10, markeredgecolor="green", markerfacecolor="red", linestyle='none')


#fetch some reasonable axes limits
xmax = ceil(1.1*max(A[0]+r_1,B[0]+r_2,C[0]+r_3))
xmin = floor(.9*min(A[0]-r_1,B[0]-r_2,C[0]-r_3))
ymax = ceil(1.1*max(A[1]+r_1,B[1]+r_2,C[1]+r_3))
ymin = floor(.9*min(A[1]-r_1,B[1]-r_2,C[1]-r_3))
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])
```
