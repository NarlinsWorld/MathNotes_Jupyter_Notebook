```python
# importing networkx module
import networkx as nx
%matplotlib inline

# creating sample graph object 'path_graph', 'complete_graph', cycle_graph,
# pertersen_graph 
# see https://networkx.org/documentation/stable/reference/generators.html
G = nx.wheel_graph(8)


# complement of G and saving in G_C
G_C = nx.complement(G)


# calling draw() to visualize the original graph
nx.draw(G, node_color='Green', with_labels=True)


# calling draw() to visualize the complement graph
nx.draw(G_C, node_color='Violet', with_labels=True)
```

```python
print(G.nodes)
print(G.edges)
```

```python
#Create graph with 5 nodes
G=nx.cycle_graph(5)
#Create adjacency matrix

# Create empty graph
G = nx.Graph()

# Add nodes
G.add_node(1)
G.add_nodes_from([2, 3])
G.add_node(4)

G.nodes()

# add edges

G.add_edge(1, 2)

# get graph info
print(nx.info(G))
# Name: 
# Type: Graph
# Number of nodes: 4
# Number of edges: 1
# Average degree:   0.5000

nx.draw(G, with_labels=True)
```

```python
# importing networkx module
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline

G = nx.petersen_graph()
subax1 = plt.subplot(121)
nx.draw(G, with_labels=True, font_weight='bold')
subax2 = plt.subplot(122)
nx.draw_shell(G, nlist=[range(5, 10), range(5)], with_labels=True, font_weight='bold')
```

```python
G = nx.dodecahedral_graph()
shells = [[2, 3, 4, 5, 6], [8, 1, 0, 19, 18, 17, 16, 15, 14, 7], [9, 10, 11, 12, 13]]
nx.draw_shell(G, nlist=shells)
```
