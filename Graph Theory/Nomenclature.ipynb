<font size=3>
Graph Theory has many technical terms that must be understood before much interesting material can be presented.  In this section, I will try to tersely define some terms and perhaps prove some simple theorems.  The idea is that I will be able to quickly look up a term or a proof using my math vocabulary.<br><br>
    
<span class='bigblue'> Vertices:</span> Vertices are the dots. Formally they are a set of labelled points. They do not have coordinates, just names. <br><br>
<span class='bigblue'>Edges:</span> An edge is a line between two points. It is a two element subset drawn from the set of vertices. In a diagram, it is the line between the two vertices.<br><br>
    
<span class='bigblue'> Simple Graph:</span> A set of vertices, some of which may or not be connected with edges. If $G$ is a graph, then
$$G=\left(\{vertex set\},\{edge set\}\right)=\left( \{v_1,v_2,\ldots v_n \}, \left\{\{v_i,v_j\},\{v_a,v_b\}\right\} \right)$$ where the subscripts on the 2-element subsets of edges refer to the named vertices. A simple graph excludes<br>
a) Loops - one edge connected to the same vertex at its two ends.<br>
b) Skeins - two edges connected to the same two vertices.<br>
c) Arrows - an edge to be traversed in one direction only.<br><br>
    
<span class='bigblue'> Null Graph:</span> A null graph is one with no edges, only vertices. Some authors also include a graph with no vertices too, but it is less common. Example: $N_6$  is a graph with $6$ vertices and no edges.<br><br>
    
<span class='bigblue'> Connected</span> When used with vertices, it just means that an edge exist between two vertices.  When used with a graph, as in "connected graph" it means that every vertex can be reached from any other vertex by following some edges.<br><br>
    
<span class='bigblue'> Adjacent</span> When used with vertices, it means that there is an edge between the two vertices. <br><br>
    
<span class='bigblue'> Isomorphic</span> Having the same vertex and edge set, even though the vertices may be randomly placed. Two graphs are isomorphic if they are the same, except that the vertices are labelled differently.<br><br>
    
<span class='bigblue'> Automorphic</span>Two graphs are automorphic if they are completely the same, including the vertex labeling.
<span class='bigblue'> Planar</span> A graph is planar if it is possible to draw in the plane without any crossed edges.<br><br>
    
<span class='bigblue'> Nonplanar</span> A graph that is impossible to draw in the plane except with one or more crossed edges.<br><br>
    
<span class='bigblue'> Utility Graph</span> The nonplanar graph derived from the utility problem. Three utility companies run wires to three houses without any wires crossing. This is sometimes abbreviated $K_{3,3}$ or $UG$<br><br>
    
<span class='bigblue'> Complete Graph</span> A simple graph in which every vertex is connected to every other vertex by an edge. These are designated as $K_1,K_2,\ldots ,K_n$  with the number being the number of vertices. Starting at $K_5$ all complete graphs are nonplanar.<br><br>
    
<span class='bigblue'> Degree</span> The degree of a vertex is a count of the number of edges coming into the vertex. The maximum degree of a graph $\Delta G,$ and the minimum degree of a graph $\delta G$ are the maximum and minimum of its vertices' degrees. <br><br>
    
<span class='bigblue'> Leaf Vertex (Pendant vertex):</span> A vertex of degree one.<br><br>
    
<span class='bigblue'> Cycle Graph</span> When meant to be designated as $C_n$, then it is a series of degree 2 vertices connected into a circle. The subscript $n$ gives the number of vertices.  In context, cycle may also refer to a circular path within a graph. A cycle path goes through any vertex one time only.<br><br>
    
<span class='bigblue'> Circuit</span> A path that begins and ends at the same vertex and may go through the same vertex multiple times.<br><br>
    
<span class='bigblue'> Complement graph</span> In order for a complement graph to exists, there must first be a graph. If $G$ is a graph, then we call $\overline{G}$ a complement if all vertices are the same, but all edges have been complemented. That is every existing edge in $G$ has been erased and for every vertex pair in $G$ that had no connection, a new edge has been constructed.<br><br>
    
<span class='bigblue'> Subgraph</span> A subgraph is a subset of a graph that can be created entirely with an eraser. Any vertex or edge may be erased, but none may be created. If any vertex is erased, then all of its edge connections must also be erased. Also, as in subsets, every graph is a subgraph of itself.<br><br>
    
<span class='bigblue'> Clique</span> A subgraph of $G$ that is a complete graph.

<span class='bigblue'> Supergraph</span> A supergraph is a new graph created from $G$ by the addition of new vertices and new edges. No existing vertices or edges may be modified. That is, the new graph must not erase any vertex, or modify any edge.  Thus, we cannot put a new vertex onto an existing edge as that would change the adjacency definition for the existing graph.  Some authors say 'no splicing' to mean 'no new vertex on top of an existing edge. Every graph is a supergraph of itself.<br><br> 

<span class='bigblue'> Expansion graph</span> A graph created from an existing graph solely by splicing new vertices into existing edges. Every expansion graph is an expansion graph of itself, meaning that 'make no change' can also be called an expansion. <br><br>
    
<span class='bigblue'> Faces or Regions</span> For a connected, planar graph, drawn in the plane without edge crossings, a face of the graph is a region bounded by a set of edges and vertices. There can be many regions within a graph, but there is one that is unbounded, known as the outer face or infinite face. A two vertex, 1 edge graph has as its only face the infinite face. <br><br>
    
<span class='bigblue'> Polygonal</span> A graph is polygonal if it is planar, connected, and every edge borders on two different faces. Thus $K_2$ in not polygonal, since the outer region borders the only edge on both sides. <br><br>
    
<span class='bigblue'> Outer planar graph</span> A graph for which every vertex lies on the infinite face.<br><br>
    
<span class='bigblue'> Walk</span> A walk is a sequence of connected vertices.<br><br>
    
<span class='bigblue'> Path</span> A path is a walk in which no node appears more than once. Path graphs are denoted as $P_n$ where the $n$ is the number of nodes in the path.<br><br>
    
<span class='bigblue'> Tree</span> A tree is a connected graph containing no cycles.<br><br>
    
<span class='bigblue'> Edge Disjoint Path</span> A path that uses completely different nodes (disjoint) from some other path to go from $A$ to $B.$<br><br>
    
<span class='bigblue'> Bipartite Path</span> A graph that can be partitioned into two vertex sets such that elements in set A connect exclusively to elements in set B and vice versa.<br><br>
    
 <span class='bigblue'> Complete Bipartite Graph</span>  A bibartite graph is one in which the nodes can be separated into two groups and the nodes of group A connected exclusively to the nodes of group B. A complete bipartite graph just means that all possible connections of that type have been made.  Clearly, a complete graph and a complete bipartite graph are different for $n\ge 3.$<br>
A complete bipartite graph can be designated as $K_{m,n}$ where $m$ indicates one group of nodes and $n$ the other. Example: $K_{1,3}$ "the claw" can be drawn as a central node surrounded by $3$ other nodes.
    
<span class='bigblue'> Regular Graph</span> A graph for which all vertices have the same degree.<br><br>
    
<span class='bigblue'> Platonic graph</span> A graph that is polygonal, regular, and every face is bounded by the same number of edges. Note that this is the opposite of 'every edge borders the same number of faces'. Get them unconfused if you can. These are named for the platonic solids. See Platonic Graphs at Wikipedia for the list of 5.  <br><br>
    
<span class='bigblue'> Adjacency Matrix</span> The vertices are listed in both columns and rows. The matrix entry at $i,j$ is the multiplicity number for the $i,j$ edge(s).
Loops are counted as one edge. Loops show up when $i=j.$ <br><br>
    
<span class='bigblue'> Incidence Matrix</span>List the nodes for the rows and the named edges for the columns. Every cell of an incidence matrix will be either $1$ or $0.$ If the graph is directed, $-1$ is into the node and $1$ is out of the node. <span class='blueital'>Many authors reverse the sign convention</span> It is a $1$ if the edge joins the node. Therefore, since a node has only two ends, every column in the matrix sums to 2.  Ever row in the matrix sums to the vertex degree. <br><br>
    
<span class='bigblue'> Decomposition</span> A decomposition of a graph $G$ is a set of edge-disjoint subgraphs of $G,$ $\{H_1,H_2,\ldots,H_n\}$ such that $\bigcup_i H_i=G$<br>
From the definition, for a set to be the decomposition of some graph, $G$ it must meet these three conditions. 1) $H_1, H_2,...H_n$ have to be subgraphs of $G.$<br>
2) $H_1\bigcup H_2 \bigcup \ldots H_n$=G$<br> 3)The decomposed components have no common edges.<br><br>

<span class='bigblue'> Graph Homomorphism</span> If $G,H$ are digraphs, then<br><br> $$h: V(G)\rightarrow V(H)\  s.t. (u,v)\in Edge(G)\Rightarrow(h(u),h(v))\in Edge(H).$$<br><br> If this property holds we say "h is a homomorphism." <br><br>
The tilde means "is adjacent to".

<span class='bigblue'>Definition: Homomorphism:</span> <span style='color: brown;'>Let $X$ and $Y$ be graphs. A mapping $f$ from $V(x)$ to $V(Y)$ is a homomorphism if $f(x)$ and $f(y)$ are adjacent in $Y$ whenever $x$ and $y$ are adjacent in $X.$  When there are no loops, then $x\sim y$ and $f(x)\ne f(y).$</span><br><br>
All isomorphisms are homomorphisms but the reverse is false.<br> If a homomorphism exists from $G$ to $H$ we write $$G\rightarrow H$$ and if no homomorphism exists, then we would write $G \nrightarrow H.$

<span class='bigblue'>Chromatic number, $\chi(G)$ </span> The minimum number of colors needed to properly color $G$ so that adjacent nodes always have a different color.  Ï‡(G) is the smallest $r$ such that $G\rightarrow K_r.$<br><br>
    
<span class='bigblue'>Clique number</span> Recall that a clique is a subgraph of $G$ which is a complete graph.  We say that a graph has clique number $n,$ denoted $\omega(G),$ if the $\mathbf{\text{largest}}$ clique of $G$ has $n$ vertices. Just as $\chi(G)$ is the $\mathbf{\text{smallest }} r$ such that $G\rightarrow K_r,$ $\omega(G)$ is the $\mathbf{\text{largest }} n$ such that $K_n \rightarrow G.$<br><br>
    
<span class='bigblue'>Core of a graph</span> The homomorphism of $G$ (homomorphism equivalence class of $G$) with the least number of vertices.<br><br>  

    
<span class='bigblue'>Antisymmetric</span> Let $R$ be a relation on a set $A.$ We say $R$ is antisymmetric iff $xRy$ and $yRx$ $\mathbf{only}$ when $x=y.$<br>
    That is, a relation is antisymmetric if no two $\underline{distinct}$ elements are symmetrically related.<br>
The 'divides' relation on a set of positive integers is antisymmetric.<br> Let $A=\{1,2,3,4,5,6\}$ Then <br>
 $$R=\{(x,y):x|y\}=\{(1,1),(2,2),\ldots ,(6,6),(1,2),(1,3),\ldots,(1,6),(2,4),(3,6)\}$$<br>
We see that $\mathbf{distinct}$ elements cannot be switched.  Element $(3,6)$ is in the set because $3$ divides $6,$ but $(6,3)$ is not included. The relation is antisymmetric.<br><br>
    A relation is just a set of ordered pairs.  It doesn't have to be derived from some special function. Therefore, we can create a relation that is antisymmetric.
    $$(1,3),(2,1),(4,5)$$ This relation is antisymmetric because it does not contain any element such that $x\ne y \land (x,y),(y,x)$ are both in the set.<br><br>
One more example. $$\{(a,b),(b,a),(a,c)\}\Leftarrow \text{ not symmetric and not antisymmetric}$$
    $$\{(a,a),(b,b)\}\Leftarrow \text{ both symmetric and antisymmetric.}$$
Symmetric requires that all elements be reversible. Antisymmetric requires that all distinct elements be not reversible.
    <br><br>
<span class='bigblue'>Poset</span> Partially Ordered Set is a set $P$ equipped with a binary relation $\le$ which is a partial order on $X,$ i.e., $\le$ satisfies the following three properties:
<ol>
    <li>If $x\in P$, then $x\le x$ in $P$ (reflexive property).</li>
<li>If $x, y, z \in P, x\le y$ in $P$ and $y\le x$ in $P,$ then $x = y$
    (antisymmetric property).</li>
<li>If $x, y, z \in P, x \le y$ in $P$ and $y\le z$ in $P,$ then
    $x\le  z$
    in $P$ (transitive property).</li>
    </ol>
<br>An example of a poset is the power set of $S$ when $S=\{1,2,3\}$
    $$P(S)=\left\{\emptyset,\{1\},\{2\},\{3\},\{1,2\},\{1,3\},\{2,3\},\{1,2,3\}\right\}$$
    $$R=\{(x,y)\in P(S)\times P(S)\,|\,x\subseteq Y\}$$ The relation reads as <br>
    "The set of ordered pairs x y where both x and y are from the power set of S.  So x and y are both subsets of S and x and y relate to each other and thus the ordered pair (x,y) are in a relation if $\underline{\text{x is a subset of y.}}$"<br>
    Then, $$\left( P(S),R\right) \text{ is a poset.}$$
    <br><br>In a poset, it is not required that all elements relate to each other in some way.  For example in $P(S),$ if we consider $\{1,2\}$ and $\{1,3\}$ neither of these subsets are subsets of each other so $P(S) has only been partially ordered.

<h2>Hasse Diagrams</h2>

Suppose we have the set $A=\left\{ 2,3,5,6,9,15,20,30\right\} $ and
the relation $R=\left\{ (x,y)\in A\times A:x\ |\ y\right\} .$ Hasse
diagrams are a technique for making drawings of partially ordered
sets. In the poset $(A,R)$ we will have the following pairs.
$$
\left\{ (2,20),(2,6),(3,6),(3,9),(3,15),(5,20),(6,30),(15,30)\right\} 
$$

There are other pairs as well that meet the condition.
$$
\left\{ (3,30),\{5,30\}\right\} 
$$
and of course, each of the members of $A$ will divide itself, so
$$
\left\{ \left\{ 2,2\right\} ,\left\{ 3,3\right\} ,\left\{ 5,5\right\} ,\left\{ 6,6\right\} ,\left\{ 9,9\right\} ,\left\{ 15,15\right\} ,\left\{ 20,20\right\} ,\left\{ 30,30\right\} \right\} 
$$

A Hasse Diagram tries to reduce all of this information into a simple
drawing.
<figure>
<img src='../Images/Hasse_Diagram_2.png' width=50%>
<figcaption>Hasse Diagram of poset $(A,R)$, where $A=\left\{ 2,3,5,6,9,15,20,30\right\} $
and $R=\left\{ (x,y)\in A\times A:x\ |\ y\right\} $ which is simply
division. The partial order is that an element of $A$ divides some
other element of $A.$ </figcaption>
</figure>
<br><br>

Posets are
<ul>
    <li> Reflexive: $aRa.$ Read "$a$ is related to $a.$" </li>
<li> Antisymmetric: if $a\le b$ and $b\le a$ then $a=b.$ The reverse
of this is read as distinct elements in an ordered pair cannot be
    switched.</li>
<li> Transitive: if $a\le b$ and $b\le c$ then $a\le c.$ </li>
    </ul>
    
In the diagram, the reflexive property is considered obvious. Every
element is a pair with itself, so nothing is shown. The crux of the
diagram is the antisymmetric property. The diagram flows (in this
case from bottom to top, but it could be the opposite, or left to
right). Order is shown by the flow. Each line leads to the next value
that the number at the bottom of the line divides. 

That is, in the diagram, $2$ divides $6$ and $2$ divides $20,$
$3\,|\,6,3\,|\,9$ and $3\,|\,15.$

But $3\,|\,30$ as well and no line connects $3$ to $30.$ That is
because of the transitive property. If we can find an upward only
path from a number, such as $3$ to another number, such as $30,$
then by transitivity, $3\,|\,30.$ 

That is the basics of these diagrams. Sometimes, as in this one, it
is possible to arrange the diagram such that no lines cross. It then
known as upward planar.

<span class='bigblue'>Maximal:</span> An element $y$ in a poset $A$ with partial order $R$ is
maximal if $y$ relates to no element except itself. In our diagram,
$20$ and $30$ are both maximal.

<br><br>

<span class='bigblue'>Minimal:</span> An element $x$ in a poset $A$ with partial order $R$ is
minimal if no element except itself relates to it. In our diagram,
$2,3,5$ are minimal.
<br><br>

<span class='bigblue'>Maximum:</span> An element $M$ is a maximum if every single element relates
to it. That is, if $(a,M)\in R$ is true for all $a\in A.$ No element
in our diagram has this property. If it existed, it would be unique.
One can write ``An element $M\in(A,R)$ is the maximum, if it is
greater than every other element in the poset. The symbol, $\succeq,$
is read 'succedes or equals'. So 
$$
M\succ a\ \forall\ a\in A.
$$
<br><br>

<span class='bigblue'>Minimum:</span> An element $m$ is a minimum if it does not relate to any
other element. That is, a vertex corresponds to the minimum element
if there is an upward path from a single minimum to all other vertices.
One can write ``An element $m\in(A,R)$ is the minimum, if it is
less than every other element in the poset. The symbol, $\prec,$
is read 'precedes . So 
$$
m\prec a\ \forall\ a\in A.
$$

<h3>Bounds of a Subset</h3>

A subset of a poset, is like any subset of any set. Just remember
that a proper subset, $\subset$ excludes equality while a subset,
$\subseteq$ includes equality. 

<span class='bigblue'>Upper Bound:</span> An element $z\in A$ is called an upper bound of subset
$S,$if $y\preceq z\ \forall\ y\in S$.
<br><br>

<span class='bigblue'>Lower Bound:</span> An element $x\in A$ is called a lower bound of subset
$S,$if $x\preceq y\ \forall y\in S.$
<br>
This isn't difficult. Think $x\preceq y\preceq z,$ which is the usual
way of writing that $x$ is the smallest, $y$ is in the middle, and
$z$ is the largest.

Be aware of the equality condition on an upper or lower bound.<br><br>

<h3>Bounds Reiterated</h3>

Another way to define the upper and lower bounds, which is equivalent
to the above and may or not be easier to comprehend.

<span class='bigblue'>Lower Bound:</span> An element $x\in A$ is a lower bound of subset $S$
if $\forall\ y\in S\ (x,y)\in R.$ That is, if $x$ is the lower bound
of subset $S,$ then $x$ must be related to every element in subset
$S.$ 

<br><br>

<span class='bigblue'>Upper Bound:</span>  An element $x\in A$ is an upper bound of subset $S$
if $\forall\ y\in S\ (y,x)\in R.$ That is, if $x$ is an upper bound
of subset $S,$ then $x$ must be related to every element in subset
$S.$ 
<br><br>
In both definitions, the key is that $x$ must be related to all of
the subset elements, either going up (upper bound) or down (lower bound).
Every element of the subset must have a directional (up for upper
and down for lower) path to the subsets bounds.
