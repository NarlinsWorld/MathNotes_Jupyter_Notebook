Given a parabola: $$y=x^2-4x+9/2 \tag*{EQ 1}$$ and an external point, $D=(2,-3/2),$ plot both and show the equation of a line through $D$ that is tangent to the parabola.
    
Get the slope from the first derivative. $$m=2x-4$$
A problem here is that the slope is not a number!!  Write the tangent line as if it were a number. $$y-y_D = (2x-4)(x-x_D) \tag*{EQ 2}$$
    
To get the tangent points, solve EQ1 and EQ2 simultaneously.
Work with EQ2 first, 
$$\begin{align}
y-y_D&=(2x-4)(x-x_D)\\
y&=2x^2-4x-2x\cdot x_D-4\cdot (-x_D)+y_D\\
y&=2x^2-4x-2x\cdot \left(2\right)-4\cdot \left(-2\right)-\frac{3}{2}\\
y&=2x^2-8x+\frac{13}{2} 
\end{align}$$
    
Now equate the right side of these two equations and solve for $x$.

$$\begin{align}
2x^2-8x+\frac{13}{2} &= x^2-4x+\frac{9}{2}\\
x^2-4x &= \frac{9}{2}-\frac{13}{2}\\
x^2-4x &= -2\\
(x-2)^2 &= 4-2\\
(x-2)^2 &= 2\\
x-2 &= \pm\sqrt{2}\\
x &= \pm\sqrt{2}+2
\end{align}$$
    
Now substitute that $x$ value into EQ1 to compute $y.$
    $$y = \frac{5}{2}$$
    $$\text{Tangent Point }= A = \left(\pm\sqrt{2}+2,\frac{5}{2}\right)$$
    
Get the slope of the tangent line from $D=\left(2,-\frac{3}{2}\right)$ to $A$
    
    
$$m = \frac{\frac{5}{2}+\frac{3}{2}}{\pm\sqrt{2}+2-2} = \frac{4}{\pm\sqrt{2}}$$
    
Now we can write the tangent lines.
$$\begin{align}
y-y_D &= m(x-x_D)\\
y-\frac{-3}{2} &= \left( \frac{4}{\pm\sqrt{2}} \right)\left( x- 2 \right)\\
y &= \left( \frac{4}{\pm\sqrt{2}} \right)\left( x- 2 \right) -\frac{3}{2}
\end{align}$$

```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Data for plotting
t = np.arange(-10.0, 10.0, 0.01)      #1st value, end value(excluded), increment
s = t**2-4*t+4.5

#fetch some reasonable axes limits
xmin = -5;  xmax = 15
ymin = -5;  ymax = 12

fig, ax = plt.subplots(figsize=(10,8))
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])
#plot the tangent points
ax.plot(np.sqrt(2)+2,5/2,marker="o", markersize=10, markeredgecolor="green", markerfacecolor="red", linestyle='none')
ax.plot(-np.sqrt(2)+2,5/2, marker="o")

#plot the external point
ax.plot(2,-3/2, marker="o")
#plot the parabola
ax.plot(t, s)

#compute and plot the tangent lines.
s=4/np.sqrt(2)*(t-2)-3/2
ax.plot(t,s)
s=-4/np.sqrt(2)*(t-2)-3/2
ax.plot(t,s)



ax.set(xlabel='x-axis', ylabel='y-axis',
       title='Tangent to a Parabola')
ax.grid()

#Plot a line segment
# xvalues = [.25,1.75]
# yvalues = [.5,1.5]
# plt.plot(xvalues, yvalues, 'bo', linestyle="--")

#plot an internal text label
#plt.text(.25, .6, 'A', fontsize = 18)

```
