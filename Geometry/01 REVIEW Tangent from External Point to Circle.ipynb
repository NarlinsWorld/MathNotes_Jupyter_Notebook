Given values for points A and C and radius r, find the tangent point P.
$$h=\sqrt{(A_x - C_x)^2 + (A_y-C_y)^2}$$
$$t = \sqrt{h^2-r^2}$$
$$\mathbf{u}=\frac{C-A}{|C-A|}=\left(\begin{array}{c}(C_x-A_x)/h \\(C_y-A_y)/h \end{array}\right)$$
$$\mathbf{v}=\left( \begin{array}{c}\mathbf{u_y}\\ \mathbf{-u_x}\end{array}  \right)$$
$$\cos\alpha = \frac{t}{h}\qquad \alpha = \cos^{-1}\left(\frac{t}{h}\right)$$
$$P = A + t\cdot \cos\alpha \cdot \mathbf{u}+t\cdot\sin\alpha\cdot\mathbf{v}$$

```python
%reset -f
from numpy import *
import matplotlib.pyplot as plt
%matplotlib inline

#Set up the plot
fig, ax=plt.subplots(figsize=(8,8))
ax.set(aspect=1)

#Center & radius Coordinates of Circle 1:
C=(4,3); r=2
circle1 = plt.Circle(C, r, color='b', fill=False)
ax.add_patch(circle1)

#fetch some reasonable axes limits
xmax = 15
xmin = -1
ymax = 10
ymin = 0
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])

A=(12,6)
#Plot the center of the circle
plt.plot(C[0],C[1],marker="o", markersize=10, markeredgecolor="green", markerfacecolor="red", linestyle='none')

#Plot the hypotenuse 'h'
plt.plot([A[0],C[0]],[A[1],C[1]],'bo',linestyle='-')


x=[A[0]]
y=[A[1]]
plt.plot(x,y,marker="o", markersize=10, markeredgecolor="green", markerfacecolor="red", linestyle='none')
plt.grid(True)

#figure out P
h = sqrt((A[0]-C[0])**2+(A[1]-C[1])**2)
t = sqrt(h**2-r**2)
u=((C[0]-A[0])/h,(C[1]-A[1])/h)
v=(u[1],-u[0])
α = arccos(t/h)
P = (A[0]+t*cos(α)*u[0]+t*sin(α)*v[0], A[1]+t*cos(α)*u[1]+t*sin(α)*v[1])

plt.plot(P[0],P[1],marker="o", markersize=10, markeredgecolor="green", markerfacecolor="red", linestyle='none')
plt.plot([A[0],P[0]], [A[1],P[1]], 'bo', linestyle="--")

#plot the radius 'r'
plt.plot([C[0],P[0]],[C[1],P[1]])

#plot text
plt.text(12.4, 6, 'A', fontsize = 18)
plt.text(C[0]-.8,C[1]-.4,'C',fontsize = 18)
plt.text(P[0],P[1]+.5,'P',fontsize=18)
plt.text(8,4,'h',fontsize=18)
plt.text(8,6-.3,'t',fontsize=18)
plt.text(3.5,3.7,'r',fontsize=18)
```
