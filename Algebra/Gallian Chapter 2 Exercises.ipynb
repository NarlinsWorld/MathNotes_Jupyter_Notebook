c. function composition of polynomials with real coefficients<br>
d. multiplication of 2x2 matrices with integer entries

  $$\left(\frac{12}{7}\div \frac{4}{6}\right)\div\frac{8}{3}=\frac{27}{8}$$
 $$\left(\frac{12}{7}\right)\div \left(\frac{4}{6}\div\frac{8}{3}  \right)=\frac{48}{7}$$<br>

<span class="proof">Problem 2.16 (Gallian):</span> Show that the set $\{5, 15, 25, 35\}$ is a group under multiplication modulo 40. What is the identity element of this group? Can you see any relationship between this group and U(8)?

    
<span class="proof">Answer:</span> <br> An identity element is one that leaves any element of the set to which it belongs unchanged when combined with it by a specified operation.  Here the operation is multiplication modulo 40.  Our identity element is $25.$<br>
 $$\begin{align}
 25\cdot 25 &\equiv 25\pmod{40}\\
  15\cdot 25 &\equiv 25\pmod{40}\\
   35\cdot 25 &\equiv 25\pmod{40}\\
  5\cdot 25 &\equiv 25\pmod{40}
 \end{align}$$
 To be a group, we need to establish associativity. The given set has an Cayley table with 16 multiplications modulo 40.  $$f(a,b)\equiv r\pmod{40}$$ In particular, the function is $$f(a)=r=a-40\left\lfloor \frac{a}{40}\right\rfloor$$  Since function composition is associative, the group is associative.
    
 Lastly, we need to show that every element has an inverse.  This is a little unusual because the identity element is not 1 but rather 25.  We need to show that some members of the set, $a,$ when multiplied by each element will get $25\bmod{ 40}.$  We actually did this when we found the identity element.
    $$\begin{align}
    a\cdot 5 &\equiv 25=e \pmod{40}\qquad a=5.\\
    a\cdot 15 &\equiv 25=e \pmod{40}\qquad a=15.\\
    a\cdot 25 &\equiv 25=3 \pmod{40}\qquad a=25.\\
    a\cdot 35 &\equiv 25=e \pmod{40}\qquad a=35.\\
    \end{align}$$
    
 The group $U(8)$ is defined to be the group of all posivite numbers less than $8$ and relatively prime to $8.$ So $$U(8)=\{ 1,3,5,7 \}.$$  We see that the problem group is just $5\cdot U{8}.$

<span class="proof">Problem 2.24 (Gallian):</span> (Socks–Shoes Property) (1)Draw an analogy between the statement $(ab)^{-1}=b^{-1} a^{-1}$ and the act of putting on and taking off your socks and shoes. <br>
(2) Find distinct nonidentity elements $a$ and $b$ from a non-Abelian group such that $(ab)^{-1}=a^{-1}b^{-1}.$ <br>
(3) Find an example that shows that in a group, it is possible to have $(ab)^{-2}\ne b^{-2}a^{-2}$ <br>
(4) What would be an appropriate name for the group property $(abc)^{-1}=c^{-1}b^{-1}a^{-1}?$

    
<span class="proof">Answer:</span><br>Part 1. By analogy, if you have a sock and shoe on one foot that is like $(ab)^{-1}$ when $a$ is a sock. When it is time to reverse them, $b^{-1},$ the sock has to come off first followed $a^{-1},$ the shoe.
    
For (2) and (3), Consider the non-Abelian group $D_4.$ We know it is not Abelian because $HR_{90}=D$ but $R_{90}H=D'.$
    <img src="..\Images\D4 group.png" width=70%>
All individual element motions begin from the state $R_0.$ So we can define the inverse of each element.<br>
$R_0^{-1}=R^0$<br>
$R_{90}^{-1}=R_{270}$<br>
$R_{180}^{-1}=R_{180}$<br>
$R_{270}^{-1}=R_{90}$<br>
$H^{-1} = H$<br>
$V^{-1} = V$<br>
$D^{-1} = D$<br>
$D'^{-1} = D'$<br>

    
Part 2. $(ab)^{-1}=a^{-1}b^{-1}.$ Let $a=H, b=R_{90}.$  <br>
$$\begin{align}
(ab)^{-1}&=(HR_{90})^{-1}\\
&=D^{-1}\\
(ab)^{-1} &=D
\end{align}$$
$$\begin{align}
a^{-1}b^{-1}&=H^{-1}R_{90}^{-1}\\
&=HR_{270}\\
a^{-1}b^{-1} &= D
\end{align}$$
So $(ab)^{-1}=a^{-1}b^{-1}.$
    
Part 3.  Now we want to find an example where $(AB)^{-2}\ne A^{-2}B^{-2}$<br>
Let $a=R_{90}, a^{-1}=R_{270}, b=D' b^{-1}=D'.$<br>
$$\begin{align}
(ab)^{-2} &= (ab)^{-1}(ab)^{-1}\\
&= (b^{-1}a^{-1})(b^{-1}a^{-1})\\
&= (D'R_{270})(D'R_{270})\\
&= VV\\
(ab)^{-2}&= R_0
\end{align}$$
$$\begin{align}
b^{-2}a^{-2} &= (b^{-1}b^{-1})(a^{-1}a^{-1})\\
&=(D'D')(R_{270}R_{270})\\
&=R_0R_{180}\\
&=R_{180}
\end{align}$$
So $(ab)^{-2}\ne b^{-2}a^{-2}.$
    
Part 4.  Should such group property exists, we might call it the nxn Invertible Matrix Theorem.<br>
By the property of any $n\times n$ invertible matrix, if $A_1,A_2,\cdots ,A_k$ are each $n\times n$ invertible matrices, then
    $$(A_1A_2\cdots A_k)^{-1}=A_k^{-1}A_{k-1}^{-1}\cdots A_{2}^{-1}A_1^{-1}$$
So if $a,b,c$ are $n\times n$ invertible square matrices, then $$(abc)^{-1}=a^{-1}b^{-1}c^{-1}$$

 $$\begin{align}

 (ab)^{-1}&=b^{-1}a^{-1} \text{  by socks and shoes}\\

 \text{so }\left((a^{-1}b^{-1})\right)^{-1} &= \left((b^{-1}a^{-1})\right)^{-1}\\

 \end{align}$$

<span class="proof">Problem 2.53(Gallian)</span> Suppose that in the definition of a group $G,$ the condition that there exists an element $e$ with the property $ae=ea=a$ for all $a$ in $G$ is replaced by $ae=a$ for all $a$ in $G.$ Show that $ea=a$ for all $a$ in $G.$ (Thus, a one-sided identity is a two-sided identity.)
    
<span class="proof">Answer </span> <br>If $ae=a,$<br> then $e=a^{-1}a.$<br> So $a(a^{-1}a)=a$<br> and $(aa^-1)a=a$ by associativity<br> Then $ea=a$ QED

<span class="proof">Problem 2.54(Gallian)</span> Suppose that in the definition of a group $G,$ the condition that for each element $a$ in $G$ there exists an element $b$ in $G$ with the property $ab = ba = e$ is replaced by the condition $ab = e.$ Show that $ba = e.$ (Thus, a one-sided inverse is a two-sided inverse.)
    
<span class="proof">Answer </span> Let $b^{-1}$ be the element s.t. $bb^{-1}=e.$
Start with $ba$ and algebraically arrive at $e.$
 $$\begin{align}
 ba &= ba\cdot e\\
 &= ba(bb^{-1})\\
 &= b(ab)b^{-1}\\
 &= beb^{-1}\\
 &= bb^{-1}\\
 &= e \tag*{$\square$}
  \end{align}$$
