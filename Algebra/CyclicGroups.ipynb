The integers mod, $n,$ under addition are always a cyclic group, $\mathbb{Z}_n$. The generators of a cyclic group are those elements within the group that will generate the whole group whenever they are raised to the power of the integers consecutively. Not all members of a cyclic group will be generators.  Under addition, raising a number to the power of $n,$ means adding the number $n,$ times. The integers, mod $n,$ under addition are $\{ 0,1,2,\cdots ,n-1\}$

The order of an element, $|\langle x \rangle |$ is the number of elements generated by $x$

```python
from numpy import *
def Z(n):
    S=linspace(0,n-1,n,dtype=int) 
    return S

n=8;L=[]
G=(Z(n))
print("G=",G)
#pick an element. Name it x
for x in range(n):
    #Begin at n and count for len(G). Raise (x**count)%n and print the result
    
    #print("")
    print("x=",x,"|",end=" ")
    for cnt in range(n,len(G)+n):
        L.append((x*cnt)%n)
    print(L,end=" ")
    S=set(L)
    print("order of L = ",len(S))
    L=[]
```
