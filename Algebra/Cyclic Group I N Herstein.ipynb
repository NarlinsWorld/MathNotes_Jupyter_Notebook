The basic idea of a cyclic group is that it can be generated by a single element.  
Example 2.2.4 Let $n$ be any integer. We construct a group of order $n$ as follows: $G$ will consist of all symbols $a^i, i = 0, 1, 2, ... , n - 1$ where we insist that $a^0 = a^n= e, a^i·a^i = a^{i+i}$ if $i + j\le  n$ and $a^i·a^i = a^{i+j-n}$ if $i + j > n.$ The reader may verify that this is a group. It is called a
cyclic group of order n.  

A geometric realization of the group in Example 2.2.4 may be achieved as follows: Let $S$ be the circle, in the plane, of radius $1,$ and let $\rho_n$ be a
rotation through an angle of $2\pi/n.$ Then $\rho_n \in A(S)$ and $\rho_n$ in $A(S)$ generates a group of order $n$, namely, $\{e,\rho_n, \rho_n^2,\ldots,\rho_n^{n-1}\}$.

```python
%reset -f
from math import *
from numpy import *
import matplotlib.pyplot as plt
t = linspace(0,2,1000) #1000 pts from 0 to 2
cir = cos(pi*t) #the x's
cir1 = sin(pi*t) #the y's
plt.figure(figsize=(4, 4))
line, = plt.plot(cir, cir1, color='blue', lw=.75)
#show()
n=12
a=linspace([0,0],[0,0],n) #n pts that are all zero
for i in range(n):
    a[i]=[cos(i*2*pi/n),sin(i*2*pi/n)]

for i in range(n):
    plt.plot(a[i][0],a[i][1],'ro',color='red')
    
p=linspace(0,0,n)
u=a[0]
for i in range(n):
    v=a[i]
    ρ = arccos(dot(u,v)/(linalg.norm(u)**2 * linalg.norm(v)**2))
    if i>n/2:
        ρ=2*pi-ρ
    print(i,ρ)
    p[i]=ρ
print('---------------------------')
for i in range(2*n):
    print((p[1]*i)%(2*pi))
```
