$$f(x)=\frac{|\sin(x^2+1)|}{x^2+1}$$
Can we show that the supemum of the function is its value at $x=0.$<br>
Let $a=\sqrt{\pi-1}.$  For $x\ge a$ we have
$$|f(x)|=\frac{|\sin(x^2+1)|}{x^2+1}\le \frac{1}{x^2+1} \le \frac{1}{a^2+1}=\frac{1}{\pi}$$
but <br> 
1. $f$ is positive decreasing on $[0,a).$<br>
2. $f(a)=0$<br>
3. $f(0)=\frac{sin\,1}{1}>\frac{1}{\pi}$
Thus $$\sup|f(x)|=f(0)$$
