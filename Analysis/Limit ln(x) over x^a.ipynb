It is true that the curve $\ln(x)$ grows more slowly than $x^a$ for all $a>0.$<br>
<span style='color: blue;'>We assert that $\forall a>0 \ln(x)<x^a$ for sufficiently large $x.$</span><br>
Proof: If $\ln(x)$ grows slower than $x^a$, then
$$\lim_{x\rightarrow \infty} \frac{\ln x}{x^a}=0$$ 
To take the limit, use L'Hospital's rule.
$$\lim_{x\rightarrow \infty} \frac{\ln x}{x^a}=\lim_{x\rightarrow \infty}\frac{\frac{1}{x}}{ax^{a-1}}=\frac{1}{xax^{a-1}}=
\frac{1}{ax^a}=0 \qquad \blacksquare$$
<br>This proof assumes you know that $\lim_{x\rightarrow\infty} x^a=\infty \quad \forall x>0.$ The statement is true because we let $x\rightarrow\infty$, but $a$ is merely a constant.
