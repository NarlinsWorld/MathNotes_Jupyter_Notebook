$$x_n=\frac{x_{n-1}}{2} + \frac{1}{x_{n-1}}$$
    A sequence of real numbers, which is deceasing and bounded below has a limit at its infimum.<br>
    Suppose that the series converges to $x$ as $n\rightarrow \infty.$<br> Then $$x_{n-1}\rightarrow x \text{ and } \frac{x_{n-1}}{2}\rightarrow \frac{x}{2} \text{ and }\frac{1}{x_{n-1}}\rightarrow\frac{1}{x} \text{ and}$$
    $$\underbrace{\frac{x_{n-1}}{2}+\frac{1}{x_{n-1}}}_{=x_n}\rightarrow\frac{x}{2}+\frac{1}{x}$$
and of course, $x_n\rightarrow x$, so
$$x=\frac{x}{2}+\frac{1}{x}$$
which solves to $x=\sqrt{2}.$
</font>

```python
x=[0];
x.append(10);
for n in range(2,15):
    val=x[n-1]/2+1/x[n-1];
    x.append(val.n());
    print(n,val.n())
```
